```python
# 파일 생성하기
f = open("새파일.txt", "w")
f.close()
```

```python
f = open("새파일2.txt", "w", encoding="utf-8") # f는 파일 객체
for i in range(1,11):
    data = f"{i}번째 줄입니다.\n"
    f.write(data)
f.close()

# 인코딩 >> 127개의 문자 표현하는 ASCII code
```

```python
f = open("./새파일2.txt", "r", encoding="utf-8")
line = f.readline()
print(line)
f.close
```

```python
f = open("./새파일2.txt", "r", encoding="utf-8")
while True:
    line = f.readline()
    if not line: break
    print(line)
    
f.close
```

```python
# readlines()
f = open("./새파일2.txt", "r", encoding="utf-8")
lines = f.readlines() # readlines는 전부 다 읽어낸다
print(lines)
f.close()
```

```python
# 파일에 새로운 내용 추가하기

f = open("./새파일2.txt", "a", encoding="utf-8")
for i in range(11, 21):
    data = f"{i}번째 줄입니다.\n"
    f.write(data)
f.close()
```

```python
# with 문과 함께 사용하기
with open("foo.txt", "w") as f:
    f.write("life is too short, you need Python") # 영문이라 딱히 깨지는게 없음
```
