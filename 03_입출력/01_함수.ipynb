```python
# 파이썬 함수의 구조
# def 함수이름(매개변수1, 매개변수2, ...):
#     수행할 문장1
#     수행할 문장2
#     수행할 문장3
# 함수를 호출할 때는 인수, 인수의 명을 파라메타 명이라고 함


def add(a, b):
    return a + b

# 괄호는 필수가 아니다.
# return 뒤에는 **하나의 표현식(expression)**만 오면 된다.
```

```python
add(10, 20)
```

```python
result = add(10, 20)
print(result)
```

```python
# 1. 일반적인 함수
# 입력값이 있고, 반환값이 있다

def add(a, b):
    result = a + b
    return result

sum = add(3, 4)
print(sum)
```

```python
# 2. 입력값이 없는 함수
# 입력값 x, 반환값 o

def say(): 
    return "hi"

a = say()
print(a)
```

```python
# 3. 반환값이 없는 함수
# 입력값 o, 반환값 x
def add (a, b):
    print(f"{a}, {b}의 합은 {a+b}입니다.")

sum = add(3, 4)
print(sum)
```

```python
# 4. 입력값도 없고, 반환값도 없는 함수
def say():
    print("hi")

a = say()
print(a)
```
