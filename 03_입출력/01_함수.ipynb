```python
# 파이썬 함수의 구조
# def 함수이름(매개변수1, 매개변수2, ...):
#     수행할 문장1
#     수행할 문장2
#     수행할 문장3
# 함수를 호출할 때는 인수, 인수의 명을 파라메타 명이라고 함


def add(a, b):
    return a + b

# 괄호는 필수가 아니다.
# return 뒤에는 **하나의 표현식(expression)**만 오면 된다.
```

```python
add(10, 20)
```

```python
result = add(10, 20)
print(result)
```

```python
# 1. 일반적인 함수
# 입력값이 있고, 반환값이 있다

def add(a, b):
    result = a + b
    return result

sum = add(3, 4)
print(sum)
```

```python
# 2. 입력값이 없는 함수
# 입력값 x, 반환값 o

def say(): 
    return "hi"

a = say()
print(a)
```

```python
# 3. 반환값이 없는 함수
# 입력값 o, 반환값 x
def add (a, b):
    print(f"{a}, {b}의 합은 {a+b}입니다.")

sum = add(3, 4)
print(sum)
```

```python
# 4. 입력값도 없고, 반환값도 없는 함수
def say():
    print("hi")

a = say()
print(a)
```

```python
# 입력값이 몇 개가 될지 모를때는 어떻게 할까?
def add_many(*args):
    result = 0 
    print(type(args)) # 튜플 > 값 수정 불가능한 것
    for i in args:
        result = result + i
    return result

result = add_many(1, 2, 3, 4, 5)
print(result)
```

```python
def add_mul(choice, *args):
    if choice == "add":
        result = 0
        for i in args:
            result = result + i

    elif choice == "mul":
        result = 1
        for i in args:
            result = result * i
     
    return result
```

```python
result = add_mul("add", 1, 2, 3, 4, 5)
print(result)
```

```python
def create_profile(**kwargs):
    print("===== 프로필 정보 =====")
    for key, value in kwargs.items():
        print(f"{key}: {value}")
```

```python
# 함수의 반환 값은 언제나 하나이다.

def add_and_mul(a, b):
    return a+b, a*b
```

```python
result = add_and_mul(3, 4)
print(result)
```

```python
def say_nick(nick):
    if nick == "학생":
        return # 을 만나면 함수를 종료시킨다.
    print(f"너의 별명은 {nick}이다")

say_nick("선생")
say_nick("학생")
```

```python
# 매개변수에 초기값 미리 설정하기

def say_myself(name, age, man=True):
    print(f"너의 이름은 {name}이다")
    print(f"너의 나이는 {age}이다")
    if man:
        print("남자입니다.")
    else:
        print("여자입니다")

say_myself("홍길동", 25)
say_myself("김영희", 30, False)
```

```python
# 함수 안에서 선언한 변수의 효력 범위

a = 1
def vartest(a):
    a = a + 1 # 이 안에서만 활동하는 a, 같은 a 이지만 지역변수이다. (이 안에서만 활동 가능)

vartest(a)
print(a)
```

```python
a = 1
def vartest():
    global a
    a = a + 1

vartest()
print(a)
```

```python
# lambda 예약어
# 함수 이름 = lambda 매개변수1, 매개변수2, ... : 수행할 문장
def add(a, b):
    return a + b

add = lambda a, b: a + b
```

```python
# 함수의 독스트림(Docstring)

def add(a, b):
    """
    두 숫자를 더하는 함수

    parameters:
    a(int, float): 첫 번째 숫자
    b(int, float): 두 번째 숫자

    returns:
    int, float: 두 숫자의 합

    """

    return a + b

print(add.__doc__)
```
