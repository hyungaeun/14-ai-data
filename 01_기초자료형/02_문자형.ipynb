```python
# 문 포매팅
# 1. 숫자 바로 대입
print("i eat %d apples" %3)

# 2. 문자열 바로 대입
print("i eat %sapples" %"five")

# 3. 숫자 값을 나타내는 변수로 대입
number = 3
print("i eat %dapples" %number)

# 4. 두개 이상의 값 넣기
number = 10
day = "three"
print("i ate %d apples, so i was sick for %s days." % (number,day))
```

### 문자열 포멧 코드
- %s: 문자열(string)
- %c: 문자1개(character)
- %d: 정수(integer)
- %f: 실수(float)
- %o: 8진수
- %x: 16진수
- %%: 문자 % 그 자체


```python
# format 함수를 사용한 포매팅
# 1. 숫자 바로 대입하기
print("i eat {0} apples.".format(3))

# 2. 문자열 바로 대입하기
print("i eat {0} apples.".format("five"))

# 3. 두개 이상의 값 넣기
number = 10
day = "three"
print("i ate {0} apples, so i was sick for {1} days.".format(number,day))
```

```python
# f 문자열 포매팅 ** 주로 쓰는 것, 이것만 알면 됨
name = "홍길동"
age = 30
print(f"나의 이름은 {name}입니다. 나이는 {age}입니다")
```

```python
# 문자열 관련 함수들

# 1. 문자 개수 세기 - count

a = "hobby"
print(a.count("b"))  #print(), len(), type()은 함수 / count()는 method(기능을 하는 역할), 객체가 있으면 그때 쓰는 함수(주체가 문자열 일때만 사용 가능 .count)

# 2. 위치 알려주기 - find
print(a.find("b"))

# 3. 위치 알려주기 - method
print(a.index("b"))

# 4. 문자열 삽입 - join
",".join("abcd")

# 5. 소문자를 대문자로 바꾸는 함수 - upper
a = "hello"
print(a.upper())

# 6. 대문자를 소문자로 바꾸는 함수 - lower
a = "HI"
print(a.lower())    #lower() 안이 아니라 앞에 a 적는거!!!

# 7. 공백 지우기 strip
a = "   hi   "
print(a)
print(a.strip())

# 8. 왼쪽 공백 지우기 lstrip
# 9. 오른쪽 공백 지우기 rstrip
print("왼쪽", a.lstrip())
print("오른쪽", a.rstrip())

# 10. 문자열 바꾸기 - replace
a = "life is too short"
print(a.replace("life", "your leg"))

# 11. 문자열 나누기 - split
a = "life is too short"     #shift + a/t + 방향키 (복사), a/t 방향키 (밑으로 내리기)
print(a.split())
```
