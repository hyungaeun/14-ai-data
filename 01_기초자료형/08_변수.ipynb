```python
a = [1, 2, 3]
b = a[:]
print(b) > b = a[:] 는 복제의 개념
print(id(a))
print(id(b))
```

```python
# 변수를 만드는 여러가지 방법
a, b = ["python", "like"]
print(a)
print(b)
# a, b = ["python", "like", "me"] 하면 틀림, 개수 같아야 함
```

```python
a, b = ["python", "like"]
print(a)
print(b)
print(type(a))
```

```python
a = b = "python"
print(a)
print(b)
print(type(a))
```
