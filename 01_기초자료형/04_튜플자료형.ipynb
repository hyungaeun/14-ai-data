```python
# 튜플은 어떻게 만들까?
a = ()
b = (1,) # 하나의 요소일 때 뒤에 , 가 필요
c = (1, 2, 3)
d = 1, 2, 3 # 괄호 생략 가능
e = ("a", "b", ("ab", "cd")) #튜플 안에 튜플 넣는것도 가능
f = 1,
g = 1

print(type(a))
print(type(b))
print(type(c))
print(type(d))
print(type(e))
print(type(f))
print(type(g))

```

```python
# 튜플 요소 값을 삭제하려 할때
a = (1, 2, "a", "b")
del a[0] # 지울 수 없음, 변경 x
```

```python
# 튜플 요소 값을 변경하려 할때
a = (1, 2, "a", "b")
a[0] = "c" # 튜플은 변경도 안됨
```

```python
a = (1, 2)
a = a * 2 
print(a)
```

```python
a = 10
print(id(a))
a = 20 # 10을 가리키던 a가 20을 가리킴 >> 재할당, 재선언 됨
print(id(a))
```

```python
a = [1, 2, 3]
print(id(a))

a.append(4) # 계속 같은 곳을 가리키고 있는데 거기에 4가 추가된거라 id(a) 주소가 동일한 것
print(a)
print(id(a))
```

```python
# 튜플에 인덱싱 하기
a = (1, 2, 3, 4)
print(a[0])
```

```python
# 튜플 슬라이싱 하기
a = (1, 2, 3, 4)
print(a[:2])
```

```python
# 튜플 더하기
a = (1, 2)
b = (3, 4)
c = a + b
print(c)
```

```python
# 튜플 곱하기
a = (3, 4)
b = a * 3
print(b)
```
