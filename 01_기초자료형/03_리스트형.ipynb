```python
# 리스트는 어떻게 만들까?
# 형식 

# a = 0, 1
a = []
b = [1, 2, 3]
c = ["life", "is", "too", "short"]
d = [1, 2, "life", "is"]
e = [1, 2, ["life", "is"]]

print(len(a))
print(len(b))
print(len(c))
print(len(d))
print(len(e))

print(type(a))
print(len(a))
```

```python
# 리스트의 인덱싱
a = ["life", "is", "too", "short"]
print(a[2])
print(a[-1])
a[0] # 이거 자체가 문자열
a[0][-1] # a 에 0에 위치한 게 life, 그 life에서 맨 뒤에 있는 e를 가져온 것
```

```python
a = [1, 2, 3]
a[0] + a[2]
```

```python
a = [1, 2, 3, ["a", "b", "c"]]
print(a[-1])
print(a[-1][1])
```

```python
# 리스트의 슬라이싱
a = [1, 2, 3, 4, 5]
a[0:1+1]
a[-2:]
```

```python
# 리스트 연산하기
# 1. 리스트 더하기

a = [1, 2, 3]
b = [4, 5, 6]
print(a + b)

# 2. 리스트 곱하기
a = [1, 2, 3]
print(a * 3)
```

```python
# 리스트의 수정과 삭제
# 1. 리스트의 값 수정하기

a = [1, 2, 3]
a[2] = 4
print(a)

# 2. 리스트의 요소 삭제하기
a = [1, 2, 3] # if 다시 a를 설정하지 않았다면 답이 [1,4] 였을 것
del a[1]
print(a)

a = [1, 2, 3, 4, 5]
del a[2:]
print(a)
```

```python
# 리스트 관련 함수 ** 중요
# 1. 리스트에 요소 추가하기 append

a = [1, 2, 3]
a.append(4)  #함수라 () 가 존재하는 것
print(a)
a.append([5,6])  #[5,6]을 한개의 요소로 집어넣음
print(a)
```

```python
# 2. 리스트 정렬 - sort 
a = [1,4,3,2]
 # 오름차순 형태로 정렬됨
a.sort()
print(a)
 # 내림차순 형태로 정렬됨
a.sort(reverse=True)  # bool 은 true or false, 기본값이 false라 true로 설정
print(a)
```

```python
# 3. 리스트 뒤집기 - reverse
a = ["a", "b", "c"]
# a.sort()
# print(a)    , ctrl + / 하면 전체가 주석처리 됨
a.reverse()
print(a)
```

```python
# 4. 인덱스 번호 - index (색인, 몇번 index를 가져 오는 것)
a = [5,4,3,2,1]
print(a.index(3))
print(a.index(2))
```

```python
# 5. 리스트에 요소 삽입 - insert

a = [1,2,3]
a.insert(0,4)
print(a)
```

```python
# 6. 리스트 요소 제거 - remove

a = [1,2,3,1,2,3]
a.remove(3)
print(3)
```

```python
# 7. 리스트 요소 꺼내기 - pop
a = [1, 2, 3]
result1 = a.pop() #결과를 꺼낸다
print(a)
print(result1)
```

```python
# 8. 요소의 개수 세기 - count
a = [1,2,3,1]
a.count(1)
```

```python
# 9. 리스트 확장 - extend
a = [1,2,3]
a.extend([4,5])
print(a)
```
