```python
print(1, bool(1))
print(0, bool(0))
print(2, bool(2))
print("Python", bool("Python"))
print(" ", bool(" "))
print("[]", bool("[]"))
print([1, 2, 3], bool([1, 2, 3]))
```

```python
if [1, 2, 3]:
    print("참")
else:
    print("거짓")
```

```python
# 논리 연산자
# 1. and 연산자

print(True and True)
print(True and False)
print(False and False)
print(False and True and True) # 하나라도 False 가 있으면 무조건 False

# or 연산자
print(True or True)
print(True or False)
print(False or False)
print(False or True or True)

# not 연산자
print(not True)
print(not False)
print(not 1)
print(not [])
```
