```python
result = 0 # 상태값을 쓰기 위해서 global 함수를 쓴다 / 아래 내용은 계산기의 logic과 같다.

def add(num):
    global result
    result = result + num
    return result

print(add(3))
print(add(4))
```

```python
class Calculator:
    def __init__(self): 
        self.result = 0 
                        
    def add(self, num):
        self.result = self.result + num # self.result += num 같은 의미임
        return self.result
        
    def sub(self, num):
        self.result -= num
        return self.result

    def mul(self, num):
        self.result *= num
        return self.result

    def div(self, num):
        self.result /= num
        return self.result

cal1 = Calculator() 
print(cal1.add(10))
print(cal1.sub(2))
print(cal1.mul(5))
print(cal1.div(4))

cal2 = Calculator() 
print(cal2.add(5))
print(cal2.mul(10))
print(cal2.sub(10))
print(cal2.div(5))
```

```python
class FourCal:
    def setdata(self, first, second):
        self.first = first
        self.second = second

    def add(self):
        result = self.first + self.second
        return result

    def sub(self):
        result = self.first - self.second
        return result

    def mul(self):
        result = self.first * self.second
        return result

    def div(self):
        result = self.first / self.second
        return result

cal3 = FourCal()
print(type(cal3))
cal3.setdata(4, 2)

print(cal3.first)
print(cal3.second)
print(cal3.add())
print(cal3.sub())
print(cal3.mul())
print(cal3.div())
```
