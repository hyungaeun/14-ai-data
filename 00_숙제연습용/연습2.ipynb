```python
# 2번

def calculate(a,b,operator):
    if operator == "+":
        print(a + b)
    elif operator == "-":
        print(a - b)
    elif operator == "*":
        print(a * b)
    elif operator == "/":
        if b == 0:
            print("0으로 나눌 수 없습니다.")
        else:
            print(a / b)
    else:
        print("잘못된 연산자 입니다.")


calculate(10, 3, "+")   # 13
calculate(10, 3, "-")   # 7
calculate(10, 3, "*")   # 30
calculate(10, 3, "/")   # 3.3333333333333335
calculate(10, 0, "/")   # 0으로 나눌 수 없습니다
calculate(10, 3, "%")   # 잘못된 연산자입니다
```

```python
# 3번

def sum_and_avg(a, b):
    i = a + b
    j = (a + b) / 2
    return i, j  # 다른 값으로 지정을 해주었기 때문에 반드시 return 필요

result = sum_and_avg(80, 90)
print(result) 

s, a = sum_and_avg(75, 95)
print(f"합계: {s}")   
print(f"평균: {a}")
```

```python
# 4번

def order_coffee(menu,size,shot):
    if size == "":
        size = "Tall"  # 다른 값으로 적어내지 않는 이상 return을 할 필요는 없음

    if shot == "":
        shot = "1"

    print(f"[주문] {menu} ({size}, {shot}샷)")

order_coffee("아메리카노", "", "")
```

```python
# 5번

def get_average(*args):
    if len(args) == 0:
        print("0")
    else:
        print(sum(args) / len(args))

get_average(80, 90, 100)        # 90.0
get_average(75, 82, 91, 68, 95) # 82.2
get_average(100)                # 100.0
get_average()                  # 0
```

```python
# 6번

def print_gugudan(dan, start, end):
    if start > end:
        print("잘못된 범위입니다.")
    else:
        if start == "" and end == "":
            for b in range(1,10):
                print(f"{dan} x {b} = {dan * b}")
        else:
            for a in range(start, end + 1):
                print(f"{dan} x {a} = {dan * a}")

print_gugudan(5, "", "")
print_gugudan(3, 3, 7)
```

```python
# 7번

def left_triangle(height):
    for i in range(1, height + 1):
        print("*" * i)


left_triangle(5)
```

```python
# 8번

def check_even_odd(num):
    if num % 2 == 0:
        print("짝수")
    else:
        print("홀수")

check_even_odd(10)
check_even_odd(7)
```

```python
# 9번

def my_abs(num):
    if num >= 0:
        print(num)
    else:
        print(-num)

my_abs(5)
my_abs(-3)
```

```python
# 10번

def find_max(a, b, c):
    if a > b:
        if a > c:
            a = a
        else:
            a = c
    else:
        if b > c:
            a = b
        else: 
            a = c
    print(a)

find_max(3, 8, 5)
find_max(10, 4, 7)  
```
