```python
class Product:
    """상품 정보를 관리하는 클래스"""
    
    def __init__(self, product_id, name, price, weight, stock):
        self.product_id = product_id
        self.name = name
        self.price = price
        self.weight = weight
        self.stock = stock
    
    def get_info(self):
        """상품 정보를 딕셔너리로 반환"""
        return {
            "상품ID": self.product_id,
            "상품명": self.name,
            "가격": self.price,
            "무게": self.weight,
            "재고": self.stock
        }
    
    def update_stock(self, quantity):
        """
        재고 수량 업데이트
        재고가 부족하면 False 반환, 성공하면 True 반환
        """
        if self.stock >= quantity:
            self.stock -= quantity
            return True
        else:
            return False
    
    def apply_discount(self, discount_rate):
        """할인율을 적용한 가격 반환 (원본 가격은 유지)"""
        return int(self.price * (1 - discount_rate))


print("=== 상품 정보 ===")
print(p1.get_info())
print(p2.get_info())
print(p3.get_info())

print("\n=== 재고 변경 ===")
p1.update_stock(-3)   # 3개 판매
p2.update_stock(5)    # 5개 입고
p3.update_stock(-10)  # 실패 (재고 부족)

print(p1.stock)
print(p2.stock)
print(p3.stock)

print("\n=== 할인 가격 ===")
print(p1.apply_discount(0.1))  # 10% 할인
print(p2.apply_discount(0.2))  # 20% 할인
print(p3.apply_discount(0.3))  # 30% 할인

```

```python
lass ShoppingCart:
    """장바구니를 관리하는 클래스"""
    
    def __init__(self):
        """장바구니 초기화"""
        self.cart = []
    
    def add_product(self, product, quantity):
        """상품 추가"""
    
        for item in self.cart:
            if item['product'].product_id == product.product_id:
                item['quantity'] += quantity
                return
        
        self.cart.append({
            'product': product,
            'quantity': quantity
        })
    
    def remove_product(self, product_id):
        """상품 제거"""
        self.cart = [item for item in self.cart if item['product'].product_id != product_id]
    
    def get_total_price(self):
        """총 금액"""
        total = 0
        for item in self.cart:
            total += item['product'].price * item['quantity']
        return total
    
    def get_total_weight(self):
        """총 무게"""
        total = 0
        for item in self.cart:
            total += item['product'].weight * item['quantity']
        return total
    
    def show_cart(self):
        """장바구니 출력"""
        print("\n=== 장바구니 목록 ===")
        for item in self.cart:
            p = item['product']
            q = item['quantity']
            print(f"{p.name} | 수량: {q} | 가격: {p.price * q}")
        print(f"\n총 금액: {self.get_total_price()}")
        print(f"총 무게: {self.get_total_weight():.2f}kg")


p1 = Product("P001", "사과", 1000, 0.5, 10)
p2 = Product("P002", "바나나", 1500, 0.3, 20)
p3 = Product("P003", "딸기", 3000, 0.2, 5)

cart = ShoppingCart()

cart.add_product(p1, 2)
cart.add_product(p2, 3)
cart.add_product(p3, 1)

cart.add_product(p1, 1)

cart.show_cart()

cart.remove_product("P002")

cart.show_cart()
```

```python
class Order:
    """주문 정보를 관리하는 클래스"""
    
    def __init__(self, order_id, cart, customer_name):
        self.order_id = order_id
        self.cart = cart
        self.customer_name = customer_name
        self.shipping_fee = 3000  # 기본 배송비
    
    def calculate_final_price(self):
        """최종 결제 금액 (상품 금액 + 배송비)"""
        product_total = self.cart.get_total_price()
        
        if product_total >= 50000:
            shipping = 0
        else:
            shipping = self.shipping_fee
        
        return product_total + shipping, product_total, shipping
    
    def print_receipt(self):
        """주문서 출력"""
        final_price, product_total, shipping = self.calculate_final_price()
        
        print("\n======================")
        print("       주문서")
        print("======================")
        print(f"주문 ID: {self.order_id}")
        print(f"고객명: {self.customer_name}")
        
        print("\n[주문 상품]")
        for item in self.cart.cart:
            p = item['product']
            q = item['quantity']
            print(f"{p.name} | 수량: {q} | 금액: {p.price * q}")
        
        print("\n----------------------")
        print(f"상품 금액: {product_total}")
        print(f"배송비: {shipping}")
        print(f"최종 결제 금액: {final_price}")
        print("======================")

        
laptop = Product("P001", "노트북", 1200000, 2.5, 10)
mouse = Product("P002", "무선마우스", 35000, 0.2, 50)
keyboard = Product("P003", "기계식키보드", 89000, 1.0, 30)

cart = ShoppingCart()

cart.add_product(laptop, 1)
cart.add_product(mouse, 2)
cart.add_product(keyboard, 1)

cart.show_cart()

order = Order("ORD20260211001", cart, "홍길동")

order.print_receipt()
```
