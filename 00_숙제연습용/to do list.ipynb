```python
todo_list = ["영화보기", "책읽기", "취업", "수영장 등록하기"]  

while True:
    print("\n--- To Do List ---")
    print("1. 목록 보기")
    print("2. 할 일 추가")
    print("3. 할 일 삭제")
    print("4. 종료")

    choice = int(input("선택: "))

    if choice == 1:
        if len(todo_list) == 0:
            print("할 일이 없습니다.")
        else:
            for i in range(len(todo_list)):
                print(f"{i+1}. {todo_list[i]}")

    elif choice == 2:
        task = input("추가할 일을 입력하세요: ")
        todo_list.append(task)
        print("추가되었습니다.")

    elif choice == 3:
        if len(todo_list) == 0:
            print("삭제할 일이 없습니다.")
        else:
            for i in range(len(todo_list)):
                print(f"{i+1}. {todo_list[i]}")
            num = int(input("삭제할 번호를 입력하세요: "))
            
            if 1 <= num <= len(todo_list):
                todo_list.pop(num-1)
                print("삭제되었습니다.")
            else:
                print("잘못된 번호입니다.")

    elif choice == 4:
        print("프로그램 종료")
        break

    else:
        print("잘못된 선택입니다.")
```

```python
import json
import os
from datetime import datetime

FILE_NAME = "todo_list.json"


def load_todos():
    if os.path.exists(FILE_NAME):
        with open(FILE_NAME, "r", encoding="utf-8") as file:
            return json.load(file)
    return []


def save_todos(todos):
    with open(FILE_NAME, "w", encoding="utf-8") as file:
        json.dump(todos, file, ensure_ascii=False, indent=2)


def show_todos(todos):
    if not todos:
        print("\n할 일이 없습니다.")
        return

    print("\n[To-Do List]")
    today = datetime.today().date()

    for i, todo in enumerate(todos, 1):
        status = "✔" if todo["done"] else "✘"
        due_text = todo["due"]

        warning = ""
        if due_text:
            due_date = datetime.strptime(due_text, "%Y-%m-%d").date()
            if due_date < today and not todo["done"]:
                warning = " [마감 지남]"

        print(f"{i}. [{status}] {todo['task']} (마감: {due_text or '없음'}){warning}")


def add_todo(todos):
    task = input("할 일 내용: ").strip()
    due = input("마감일 (YYYY-MM-DD, 없으면 엔터): ").strip()

    if not task:
        print("할 일 내용은 비워둘 수 없습니다.")
        return

    if due:
        try:
            datetime.strptime(due, "%Y-%m-%d")
        except ValueError:
            print("날짜 형식이 올바르지 않습니다. 예: 2026-04-01")
            return
    else:
        due = ""

    todos.append({
        "task": task,
        "done": False,
        "due": due
    })

    save_todos(todos)
    print("할 일이 추가되었습니다.")


def complete_todo(todos):
    if not todos:
        print("완료 처리할 할 일이 없습니다.")
        return

    show_todos(todos)

    try:
        num = int(input("완료 처리할 번호: "))
        if 1 <= num <= len(todos):
            todos[num - 1]["done"] = True
            save_todos(todos)
            print("완료 처리되었습니다.")
        else:
            print("유효한 번호를 입력하세요.")
    except ValueError:
        print("숫자를 입력하세요.")


def delete_todo(todos):
    if not todos:
        print("삭제할 할 일이 없습니다.")
        return

    show_todos(todos)

    try:
        num = int(input("삭제할 번호: "))
        if 1 <= num <= len(todos):
            removed = todos.pop(num - 1)
            save_todos(todos)
            print(f"삭제됨: {removed['task']}")
        else:
            print("유효한 번호를 입력하세요.")
    except ValueError:
        print("숫자를 입력하세요.")


def main():
    todos = load_todos()

    while True:
        print("\n===== TO-DO LIST =====")
        print("1. 할 일 추가")
        print("2. 목록 보기")
        print("3. 완료 처리")
        print("4. 삭제")
        print("5. 종료")

        choice = input("선택: ").strip()

        if choice == "1":
            add_todo(todos)
        elif choice == "2":
            show_todos(todos)
        elif choice == "3":
            complete_todo(todos)
        elif choice == "4":
            delete_todo(todos)
        elif choice == "5":
            print("프로그램을 종료합니다.")
            break
        else:
            print("1~5 중에서 선택하세요.")


if __name__ == "__main__":
    main()
```
