```python
# 문제 1

score = int(input("0 ~ 100 사이 점수를 입력하세요."))

while score > 100 or score <0:
    score = int(input("점수를 다시 입력하세요."))

# 현재 코드의 핵심 문제는 잘못된 점수를 다시 입력받아도 등급 계산이 다시 실행되지 않는다는 점이다.
# 즉, 재입력은 받지만 grade가 정의되지 않을 수 있다.
# if score > 100 or score < 0 >> 로 푸는 경우

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print(f"당신의 등급은 {grade} 입니다.")
```

```python
# 문제 2

number = int(input(" 2 ~ 9 중의 숫자를 입력하시오."))

if number < 2 or number > 9:
    print(" 숫자를 다시 입력하시오.")
else:
    for a in range(2, number + 1):
        for b in range(1, 10):
            print(f"{a} x {b} = {a * b}")
        print(" ")
```

```python
# 문제 2

number = int(input("몇 단을 출력하시겠습니까?")) 

if number > 9 or number < 2:
    print("범위를 벗어났습니다.")
else:
    print(f"몇 단을 출력하시겠습니까? {number}")
    for b in range(1,10):
        print(f"{number} * {b} = {number * b}")
```

```python
# 문제 3

n = int(input("숫자를 입력하시오."))
result = " "
total = 0

if n < 0 :
    n = int(input("숫자를 입력하시오."))
else:
    for a in range(1, n + 1):  # for 문은 (1, n+1까지 될때까지 자동으로 1씩 더해지면서 진행됨)
         if a == n:
             result += f"{a}"
             total += a
         else:
             result += f"{a} + "
             total += a
         
print(f"{result} = {total}")
```

```python
# 문제 3 > 2씩 늘어나게 만들고 싶은 경우

n = int(input("숫자를 입력하시오."))
result = " "
total = 0

if n < 0 :
    n = int(input("숫자를 입력하시오."))
else:
    for a in range(1, n + 1, 2):  # for 문은 (1, n+1까지 될때까지 자동으로 1씩 더해지면서 진행됨)
         if a == n:
             result += f"{a}"
             total += a
         else:
             result += f"{a} + "
             total += a
         
print(f"{result} = {total}")
```

```python
# 문제 4

홀수 = ""
짝수 = ""

for a in range(1,21):
    if a % 2 == 0:
        짝수 += f"{a}, "
    else:
        홀수 += f"{a}, "

짝수 = 짝수.rstrip(", ")
홀수 = 홀수.rstrip(", ")
print(f"{짝수}(개수:{len(짝수.split(", "))})")
print(f"{홀수}(개수:{len(홀수.split(", "))})")
```

```python
# 문제 4

홀수 = []
짝수 = [] # append는 리스트에서만 사용이 가능하다!!!

for a in range(1,21):
    if a % 2 == 0:
        짝수.append(a)
    else:
        홀수.append(a)

print(f"{짝수} (짝수: {len(짝수)}개 이다.)")
print(f"{홀수} (짝수: {len(홀수)}개 이다.)")

```

```python
# 문제 4

홀수 = []
짝수 = []

for a in range(1, 21):
    if a % 2 == 0:
        짝수.append(a)
    else:
        홀수.append(a)
print(짝수)
print(f"{', '.join(map(str, 짝수))} (짝수: {len(짝수)}개 이다.)")
print(f"{', '.join(map(str, 홀수))} (홀수: {len(홀수)}개 이다.)")

# map > [1, 2, 3] >> 1, 2, 3 인 리스트 문자열로 변환
# map(함수, 리스트) > 함수: str 가 문자열로 변환해 준것
# join > 문자열들을 하나로 합침
```
